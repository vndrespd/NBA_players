## Análisis decriptivo




URL: https://www.kaggle.com/datasets/justinas/nba-players-data



**PLAYER_NAME:**  Nombre(+apellido) de los jugadores (String, Cualilativo) 

**TEAM_ABBREVIATION:** Abreviacion del equipo del jugador (String, Cualitativo)

**AGE:** Edad  del jugador dentro de la temporada cogida (Float, Cuantitativo)

**PLAYER_HEIGHT:** Altura en cm de los jugadores (Float, Cuantitativo)

**PLAYER_WEIGHT:** Peso en kg de los jugadores (Float, Cuantitativo)


**COLLEGE** Nombre de la univerdad al que jugador fue (string, Cualitativo)


**COUNTRY** País del jugador donde nació, (String, Cualitativo)


**DRAFT_YEAR** Año en que se drafteo a ese jugador (string, Cuantitativo)


**DRAFT_ROUND** Ronda del draft en la jugador fue elegido (String, Cualitativo)

**DRAFT_NUMBER:** Nº eleccion de la ronda del draft del jugador (String,Cualitativo)

**GP:** Partidos  jugados en la temporada (Int, Cuantitativo)

**PTS:** Media de  puntos por partidos (Float, Cuantitativo)

**REB**: Media de rebotes por partido (Float, Cuantitativo)

**AST:** Media de asistencias por partido (Float, Cuantitativo)

**NET_RATING:**: Diferencia entre ratio ofensivo y ratio defensivo (Float, Cuantitativo)

**OREB_PCT:**: Porcentaje de rebote ofensivo respecto al total del equipo (Float, Cuantitativo)

**DREB_PCT:** Porcentaje de rebote defensivo respecto al total del equipo (Float, Cuantitativo)

**USG_PCT:** Porcentaje de uso ofensivo, las veces que un jugador actua en la ofensiva  (Float, Cuantitativa)

**TS_PCT:** Porcentaje de tiro verdadero, eficiencia al tirar el balon (de todos los pts) (Float, Cuantitativo)

**AST_PCT:** Porcentaje de asistencias respecto al total del equipo (Float, Cuantitativo)

**SEASON:** Temporada del jugador en la que se recogen los anteriores datos (String, Cualitativo)

## PUNTOS

```python
#vemos las descripciones de nuestros datos: la media total, max y min cuartiles, etc..
descripcion = tablaNBA["pts"].describe()
print(descripcion)
```

```python
#formulas para hayar iqr, el maximo y minimo que determinan los outliers
iqr = descripcion["75%"] - descripcion["25%"]
print(iqr)

miniqr = descripcion["25%"] - (1.5*iqr)
print(miniqr)

maxiqr = descripcion["75%"] + (1.5*iqr)
print(maxiqr)
               
```

```python
#TODOS LOS OUTLIERS MAXIMOS
outliersmax = tablaNBA[tablaNBA["pts"] > maxiqr]
# outliersmin = [tablanba["pts"]< -8.25]

display(outliersmax)
# display(outliersmin)
```

```python
outliersmin = [tablaNBA["pts"]< miniqr]
display(outliersmin)
```

```python
#maxpts es una tabla de puntos agrupadas por nombres,ya que hay nombres repetidos segun su temporada
maxpts = tablaNBA.groupby("pts")["player_name"].apply(list)
display(maxpts)
```

```python
# maxpts recogera el maximo anotado por cada nombre
maxpts = tablaNBA.groupby("player_name").max()
maxpts
```

```python
#hacemos que los pts esten ordenados siempre de mayor a menor
maxpts = maxpts.sort_values('pts', ascending=False)
```

```python
maxpts["pts"] # ordenada
maxpts["pts"][:20]
```

```python
#mayores puntos por partido pero CON OUTLIERS 
import plotly.express as px

fig = px.line(maxpts[:20], y="pts", title='higher points per game')#pone los nombres directo porque 
fig.show()
```

```python
#para hacer la grafica SIN OUTLIERS, creamos una variable que diga que mxpts muestre los puntos menores a mi  maximo tipico establecidos
sinoutliers = maxpts[maxpts["pts"] < maxiqr]
```

```python
#grafica de anotadores, pero SIN OUTLIERS,tambien hya etiquetas que te muestran la edad y el equipo de los jugadores
import plotly.express as px

fig = px.bar(sinoutliers [:20], y="pts", title='highests points per game', color='pts', text_auto="pts", hover_data=['age', 'team_abbreviation'],
          labels = {'team_abbreviation:equipo'}, height=600 )
                 
fig.show()
```
